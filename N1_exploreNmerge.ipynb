```python
df1 = pd.read_csv('c15.csv')
df2 = pd.read_csv('c16.csv')
df3 = pd.read_csv('c17.csv')
df4 = pd.read_csv('c18.csv')
df5 = pd.read_csv('c19.csv')
df6 = pd.read_csv('c20.csv')
df7 = pd.read_csv('c21.csv')
df8 = pd.read_csv('c22.csv')
offence_codes = pd.read_excel('rmsoffensecodes.xlsx')
```

```python
df_shape = [df1.shape, df2.shape, df3.shape, df4.shape, df5.shape, df6.shape, df7.shape, df8.shape]
```

```python
# Counting total number of rows of all the dataframes
li = []
total_rows = 0
for i in range(len(df_shape)):
    
    for j in range(len(df_shape[i])):
        li.append(df_shape[i][j])
for i in range(0,len(li),2):
    total_rows += li[i]
print('There are {} rows when all dataset combined'.format(total_rows))
```
